### Animais Peçonhentos

```python
df = pd.read_csv(f'{path}/ANIMBR07.csv', encoding='ISO-8859-1')

df.columns
```

```python
columns_to_keep = ['DT_NOTIFIC', 'SG_UF_NOT']
df = df[columns_to_keep]
df = df.query('SG_UF_NOT != 0')
df
```

```python
uf_dict = {
    11: 'RO',	
    12: 'AC',	
    13: 'AM',	
    14: 'RR',	
    15: 'PA',	
    16: 'AP',	
    17: 'TO',	
    21: 'MA',	
    22: 'PI',	
    23: 'CE',	
    24: 'RN',	
    25: 'PB',	
    26: 'PE',	
    27: 'AL',	
    28: 'SE',	
    29: 'BA',	
    31: 'MG',	
    32: 'ES',	
    33: 'RJ',	
    35: 'SP',	
    41: 'PR',	
    42: 'SC',	
    43: 'RS',	
    50: 'MS',	
    51: 'MT',	
    52: 'GO',	
    53: 'DF'
}

primary_dict = defaultdict(lambda: ['Year', 'State', np.zeros(12, dtype=int)])

for row in df.to_dict(orient='records'):
    date = row['DT_NOTIFIC'].split('-')
    index = date[0] + '-' + uf_dict[row['SG_UF_NOT']]
    primary_dict[index][0] = int(date[0])
    primary_dict[index][1] = uf_dict[row['SG_UF_NOT']]
    primary_dict[index][2][int(date[1]) - 1] += 1
```
